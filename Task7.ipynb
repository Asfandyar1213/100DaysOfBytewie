```python
import matplotlib.pyplot as plt

# Sample data 
time = [1, 2, 3, 4, 5]
values = [2, 5, 7, 10, 12]

# Create the line plot
plt.plot(time, values, marker='o', linestyle='-')

# Customize the plot
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Line Plot of Values over Time')
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data 
categories = ['Cat', 'Dog', 'Bird', 'Fish']
counts = [5, 3, 2, 1]

# Create the bar chart
plt.bar(categories, counts, color=['blue', 'green', 'orange', 'purple'])

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Frequency of Categories')
plt.xticks(rotation=45)  # Rotate category labels for better readability

# Display the plot
plt.show()
```
